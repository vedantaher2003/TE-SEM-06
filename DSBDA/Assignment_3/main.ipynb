```python
def get_count(df_column):
    return len(df_column)

def get_mean(df_column):
    return sum(df_column)/len(df_column)

def get_std(df_column):
    mean = get_mean(df_column)
    total = 0
    for element in df_column:
        total += (element-mean)**2
    return (total/len(df_column))**0.5

def get_min(df_column):
    return min(df_column)

def get_max(df_column):
    return max(df_column)

def get_quantile(df_column, percentile):
    n = percentile/100
    idx = int(len(df_column)*n)
    df_column = sorted(df_column)
    return df_column[idx]

def get_stats(df_column):
    print('Count : ', get_count(df_column))
    print('Mean : ', get_mean(df_column))
    print('Std : ', get_std(df_column))
    print('Min : ', get_min(df_column))
    print('25% : ', get_quantile(df_column, 25))
    print('50% : ', get_quantile(df_column, 50))
    print('75% : ', get_quantile(df_column, 75))
    print('Max : ', get_max(df_column))
```

```python
get_stats(gp_by_species.get_group("Iris-virginica")["SepalLengthCm"])
```
