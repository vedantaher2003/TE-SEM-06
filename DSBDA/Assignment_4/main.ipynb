# Assignment 3 - Data Visualization I

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic.csv")
```

```python
# Feature Explanation

# PassengerId: Unique identification number for each passenger.
# Survived: Survival status (0 = No, 1 = Yes).
# Pclass: Ticket class (1 = 1st, 2 = 2nd, 3 = 3rd).
# Name: Passenger's name.
# Sex: Passenger's gender.
# Age: Passenger's age.
# SibSp: Number of siblings/spouses aboard.
# Parch: Number of parents/children aboard.
# Ticket: Ticket number.
# Fare: Passenger fare.
# Cabin: Cabin number.
# Embarked: Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton).

df.head()
```

```python
# Line PLot - A line plot is used to visualize the relationship between two continuous variables over a continuous interval or time period.

sns.lineplot(data=df, x="Age", y="Fare")
plt.show()
```

```python
sns.lineplot(data=df, x="Age", y="Fare", hue="Sex")
plt.show()
```

```python
sns.lineplot(data=df, x="Age", y="Fare", hue="Pclass")
plt.show()
```

```python
sns.lineplot(data=df, x="Age", y="Fare", hue="Embarked")
plt.show()
```

```python
sns.lineplot(data=df, x="Age", y="Fare", hue="Survived")
plt.show()
```

```python
# Scatter Plot - A scatter plot is used to visualize the relationship between two continuous variables. Each point represents an observation in the dataset.

sns.scatterplot(data=df, x='Age', y='Fare')
plt.show()
```

```python
sns.scatterplot(data=df, x='Age', y='Fare', hue="Sex")
plt.show()
```

```python
sns.scatterplot(data=df, x='Age', y='Fare', hue="Pclass")
plt.show()
```

```python
sns.scatterplot(data=df, x='Age', y='Fare', hue="Embarked")
plt.show()
```

```python
sns.scatterplot(data=df, x='Age', y='Fare', hue="Survived")
plt.show()
```

```python
# Box Plot - A box plot is used to visualize the distribution of a continuous variable, showing the median, quartiles, and potential outliers.

sns.boxplot(data=df, x="Age")
plt.show()
```

```python
sns.boxplot(data=df, x="Fare")
plt.show()
```

```python
# Count Plot - A count plot is used to visualize the count of observations in each category of a categorical variable.

sns.countplot(data=df, x='Pclass', hue='Survived')
plt.show()
```

```python
sns.countplot(data=df, x='Sex', hue='Survived')
plt.show()
```

```python
sns.countplot(data=df, x='Embarked', hue='Survived')
plt.show()
```

```python
# Hist Plot - A histogram is a type of chart that shows the frequency distribution of data points across a continuous range of numerical values.

sns.histplot(data=df, x='Fare')
plt.show()
```

```python
sns.histplot(data=df, x='Age')
plt.show()
```
